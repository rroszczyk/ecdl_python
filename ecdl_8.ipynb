```python
# przygotowanie katalogu roboczego do pracy :)
import requests

def pobierzPlik(url, nazwaPliku):
    req = requests.get(url, stream=True)
    with open(nazwaPliku, 'wb') as uchwytPliku:
        for chunk in req.iter_content():
            uchwytPliku.write(chunk)

nazwyPlikow = ['alice.txt', 'siddhartha.txt', 'moby_dick.txt', 'little_women.txt', 'milionCyfrLiczbyPi.txt']
for nazwa in nazwyPlikow:
  pobierzPlik("https://github.com/rroszczyk/Python/raw/master/Pliki/"+nazwa, nazwa)
```

```python
nazwaPliku = 'liczby.txt'

liczby = [1, 2, 3, 5, 7, 11, 13, 17]

s = ''
for liczba in liczby:
  s = s + f'{liczba},'

with open(nazwaPliku, 'w') as uchwytPliku:          # 'w' - write - do pliku można pisać
  uchwytPliku.write(s)
```

```python
nazwaPliku = 'liczby.txt'

liczby = []

with open(nazwaPliku, 'r') as uchwytPliku:          # 'r' - read - z pliku można czytać
  napis = uchwytPliku.read()

for s in napis.split(','):
  try:
    liczby.append(int(s))
  except:
    pass

print(liczby)
```

```python
import json

liczby = [1, 2, 3, 5, 7, 11, 13, 17]
kwadraty = {}

for liczba in liczby:
  kwadraty[liczba] = liczba ** 2

print(kwadraty)

with open('liczby.json', 'w') as uchwytPliku:
  json.dump(kwadraty, uchwytPliku)
```

```python
with open('liczby.json', 'r') as uchwytPliku:
  liczby = json.load(uchwytPliku)

print(liczby)
```

```python
user_name = "Radosław"

with open('user.json', 'w') as uchwytPliku:
  json.dump(user_name, uchwytPliku)
```

```python
def wczytaj_imie_z_pliku():
  try:
    with open('user.json', 'r') as uchwytPliku:
      imie = json.load(uchwytPliku)
  except:
    return None
  else:
    return imie

def wczytaj_imie_od_uzytkownika():
  imie = input("Jak masz na imię? ")
  with open('user.json', 'w') as uchwytPliku:
    json.dump(imie, uchwytPliku)  
  return imie

imie = wczytaj_imie_z_pliku()
if imie:
  print(f"Witaj ponownie {imie}") 
else:
  imie = wczytaj_imie_od_uzytkownika()
  print(f"Witaj {imie}, zapamiętałem Cię na przyszłość")   
```

```python
def licznik_slow(nazwa_pliku):
  with open(nazwa_pliku) as plik:
    tekst = plik.read()
  slowa = tekst.split(' ')
  return len(slowa)

print(licznik_slow('alice.txt'))
```

```python
def licznik_czestosci_slow(nazwa_pliku):
  with open(nazwa_pliku) as plik:
    tekst = plik.read()
  slowa = tekst.split(' ')

  slownik = {}

  for slowo in slowa:
    if len(slowo) > 3:
      if slowo in slownik:
        slownik[slowo] += 1
      else:
        slownik[slowo] = 1

  slownik = dict(sorted(slownik.items(), key = lambda kv: (kv[1], kv[0]), reverse=True))

  print(slownik)

licznik_czestosci_slow('alice.txt')
```
